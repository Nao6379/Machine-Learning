# 動作状態での実験用動画作成
    赤 : 次に進む位置
    黒 : 現在の位置
    青 : 一個前の位置

```python
import tkinter as tk
import random
import math

# ウィンドウのサイズ
WINDOW_WIDTH = 400   #1800
WINDOW_HEIGHT = 400  #1000

# 点の初期位置
x = WINDOW_WIDTH // 2
y = WINDOW_HEIGHT // 2

# カウンタ(二個前のものを消す用)
cnt = 1
before = [x,y]

# 描画範囲の中心点
# 円の範囲を超えないように新しい点とこの点との距離を計算して円の半径(limite_value)以内になるように点を取る
x_state = WINDOW_WIDTH // 2
y_state = WINDOW_HEIGHT // 2

# 点の動く距離の大きさ
length = 50
# 円の半径
cicle_value = 100

# 点を動かすプログラム
def move_point():
    global x, y, cnt, before, after
    
    if cnt >= 2:
        # 描画したものを消して一個前,二個前のもをプロットし直す(三個前を消す)
        canvas.delete('all')
        canvas.create_oval(100,300,300,100)
        canvas.create_oval(before[0]-5, before[1]-5, before[0]+5, before[1]+5, fill='blue', tags='point')
        canvas.create_oval(x-5, y-5, x+5, y+5, fill='black', tags='point')
        canvas.create_line(before,after)
    
    # 現在の点
    before = [x,y]
    
    # ランダムな角度を生成
    angle = random.uniform(0, 2*math.pi)
    
    # 新しい位置を計算（極座標を直交座標に変換）
    new_x = x + length * math.cos(angle)
    new_y = y + length * math.sin(angle)

    # 新しい位置が単位円の内側にあるか確認して調整
    conf_x = new_x - x_state
    conf_y = new_y - y_state
    distance = math.sqrt(conf_x**2 + conf_y**2)
    
    if distance <= cicle_value:
        x, y = new_x, new_y
    else:
        # 点が円の外側に出る場合は,反対方向に射影する
        new_x = x - length * math.cos(angle)
        new_y = y - length * math.sin(angle)
        x, y = new_x, new_y
    
    # 次に進む点
    after = [x,y]
    
    #canvas.delete('point')
    canvas.create_line(before,after,tags='line')
    canvas.create_oval(x-5, y-5, x+5, y+5, fill='red', tags='point')
    
    cnt = cnt+1
    
    # 次の移動をスケジュール(数字は次の移動をどれくらい後に行うかの数字.つまり速度を変えるにはここ)
    root.after(500, move_point)
    
    
# メインウィンドウを作成
root = tk.Tk()
root.title('Random Moving Point')

# 描画領域を作成
canvas = tk.Canvas(root, width=WINDOW_WIDTH, height=WINDOW_HEIGHT, bg='white')
canvas.pack()

# 初期点を描画
canvas.create_oval(x-5, y-5, x+5, y+5, fill='black', tags='point')
canvas.create_oval(100,300,300,100)

# 点の移動を開始
root.after(3000, move_point)
#move_point()

# イベントループを開始
root.mainloop()
```
